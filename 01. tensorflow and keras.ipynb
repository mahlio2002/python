```python
import tensorflow as tf

g = tf.random.Generator.from_seed(2020)
X = g.normal(shape=(10,))
Y = 3 * X - 2

print('X: ', X.numpy())
print('Y: ', Y.numpy())
```

```python
# Loss function
def calMse(X, Y, a, b):
  YPred = a * X + b
  squaredError = (YPred - Y) ** 2
  meanSquaredError = tf.reduce_mean(squaredError)

  return meanSquaredError
```

```python
a = tf.Variable(0.0)
b = tf.Variable(0.0)

EPOCHES = 200

for epoch in range(1, EPOCHES+1):
  with tf.GradientTape() as tape:
    mse = calMse(X, Y, a, b)

  grad = tape.gradient(mse, {'a':a, 'b':b})
  dA, dB = grad['a'], grad['b']

  a.assign_sub(dA * 0.05)
  b.assign_sub(dB * 0.05)

  if epoch % 20 ==0:
    print("EPOCH %d - MSE: %.4f --- a: %.2f --- b: %.2f" %(epoch, mse, a, b))
```
